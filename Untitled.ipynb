```python
def univariate_numerical(df, feature_name,n_bins, hue_col=False):
    
    column_data = df[feature_name]
    d = pd.DataFrame(column_data.describe().round(2))
    
    font_size = 12
    fig, (ax1,ax2,ax3) = plt.subplots(1,3, gridspec_kw={'width_ratios':[2.5,2.5,1]})
    fig.set_figwidth(15)
    
    if hue_col:
        sns.histplot(data_df,x=feature_name, bins = n_bins, ax=ax1, hue=hue_col)
        sns.boxplot(data_df,x=feature_name, ax=ax2, hue=hue_col)
    else:
        sns.histplot(column_data, bins = n_bins, ax=ax1)
        sns.boxplot(column_data, ax = ax2)
    ax1.set_title(f"Histogram - {feature_name}")
    ax2.set_title(f"Boxplot - {feature_name}")
    ax3.axis('off')
    s_table = ax3.table(cellText=d.values, colLabels=d.columns, rowLabels = d.index, loc = 'center')
    s_table.auto_set_font_size(False)
    s_table.set_fontsize(font_size)
    s_table.scale(1, 1.5)
    plt.show()
```

#### Model Building

```python
acid_code = dict(zip(amino_acids,range(1,len(amino_acids)+1)))
print(acid_code)
```

```python
train_df = train.query("seq_len <= 221")
train_df.shape
```

```python
test_df = test[['seq_id','protein_sequence']].copy()
test_df.shape
```

```python
train_df['sequence'] = train_df['protein_sequence'].apply(lambda y :[acid_code.get(x,0) for x in y])
test_df['sequence'] = test_df['protein_sequence'].apply(lambda y :[acid_code.get(x,0) for x in y])
```

```python
from sklearn.preprocessing import MinMaxScaler

mm = MinMaxScaler()
y_scaled = mm.fit_transform(train_df['tm'].values.reshape(-1,1))
print(y_scaled)
```

```python
from sklearn.model_selection import train_test_split

X = train_df['sequence'].values
y = y_scaled.flatten()

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.30,random_state=123)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
test_array = test_df['sequence'].values
```

```python
print(X_train[0]), print(test_array[0])
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences

# padding sequences
max_length = 221
train_pad = pad_sequences(X_train, maxlen=max_length, padding='post', truncating='post')
val_pad = pad_sequences(X_test, maxlen=max_length, padding='post', truncating='post')
test_pad = pad_sequences(test_array, maxlen=max_length, padding='post', truncating='post')

train_pad.shape, val_pad.shape, test_pad.shape
```

```python
from tensorflow.keras.utils import to_categorical

# One hot encoding of sequences
train_ohe = to_categorical(train_pad)
val_ohe = to_categorical(val_pad)
test_ohe = to_categorical(test_pad)

train_ohe.shape, val_ohe.shape, test_ohe.shape
```

```python
from tensorflow.keras import Input
from tensorflow.keras.layers import Embedding
from tensorflow.keras.layers import Bidirectional, Dropout, Dense, LSTM
from tensorflow.keras.regularizers import l2
from tensorflow.keras.models import Model
from tensorflow.keras.callbacks import EarlyStopping
```

```python
x_input = Input(shape=(221,))
emb = Embedding(21, 256, input_length=max_length)(x_input)
bi_rnn = Bidirectional(LSTM(128, kernel_regularizer=l2(0.01), recurrent_regularizer=l2(0.01), bias_regularizer=l2(0.01)))(emb)
x = Dropout(0.1)(bi_rnn)

# relu regressor
x_output = Dense(1, activation='sigmoid')(x)

model1 = Model(inputs=x_input, outputs=x_output)
model1.compile(optimizer='adam', loss='mse')
model1.summary()
```

```python
# Early Stopping
es = EarlyStopping(monitor='val_loss', patience=3, verbose=1)

history1 = model1.fit(
    train_pad, y_train,
    epochs=50, batch_size=256,
    validation_data=(val_pad, y_test),
    callbacks=[es])
```

```python
def plot_history(history):
    loss = history.history['loss']
    val_loss = history.history['val_loss']
    x = range(1, len(loss) + 1)
    plt.plot(x, loss, 'b', label='Training loss')
    plt.plot(x, val_loss, 'r', label='Validation loss')
    plt.title('Training and validation loss')
    plt.legend()
```

```python
plot_history(history1)
```

```python
prediction = model1.predict(test_pad)
print(prediction)
```
